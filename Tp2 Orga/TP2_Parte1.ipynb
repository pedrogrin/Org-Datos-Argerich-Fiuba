#Bar Plot:

```python
df["habitat"].unique()
```

```python
df_habitat = df.groupby(["habitat", "class"]).agg({"habitat": "count"})
```

```python
df_habitat = df_habitat.rename({"habitat" : "cantidad"}, axis=1).reset_index()
```

```python
df_habitat["habitat"] = df_habitat["habitat"].replace({"g" : "grasses", "l" : "leaves", "m" : "meadows", "p" : "paths", "h" : "heaths",
"u" : "urban", "w" : "waste" , "d" : "woods" })
```

```python
df_habitat["class"] = df_habitat["class"].replace({"e" : "edible" , "p" : "poisonous"})
```

```python
df_habitat
```

```python
plt.figure(dpi=120)
sns.barplot(data=df_habitat, x="habitat", y="cantidad", hue = "class", palette=["green", "red"])
plt.title("Mushrooms amount by habitat", fontweight= "bold")
plt.xlabel("Habitat", fontweight= "bold")
plt.ylabel("Mushrooms amount", fontweight= "bold")
plt.legend(title="Mushroom class", loc='upper right')
```

#Violin Plot y Box Plot:

```python
df_cap = df[["cap-diameter", "cap-shape", "cap-surface", "class"]]
```

```python
df_cap["class"] = df_cap["class"].replace({"e" : "edible" , "p" : "poisonous"})
```

```python
df_cap["cap-surface"] = df_cap["cap-surface"].replace({ "i" : "fibrous", "g" : "grooves", "y" : "scaly", "s" : "smooth",
"h" : "shiny", "l": "leathery", "k" : "silky", "t" : "sticky", "w" : "wrinkled", "e" : "fleshy"})
#d no tiene nombre en el documento de referencias
```

```python
df_cap["cap-shape"] = df_cap["cap-shape"].replace({"b" : "bell", "c" : "conical", "x" : "convex", "f" : "flat",
"s" : "sunken", "p" : "spherical", "o" : "others"})
```

```python
df_cap
```

```python
plt.figure(dpi=120)
sns.violinplot(data=df_cap, x="cap-surface", y="cap-diameter", hue = "class", split = True, palette=["red", "green"])
plt.title("Mushrooms class by cap surface and cap diameter", fontweight= "bold")
plt.xlabel("Cap surface", fontweight= "bold")
plt.ylabel("Cap diameter", fontweight= "bold")
plt.ylim(0, 30)
plt.xticks(rotation=90)
plt.legend(title="Mushroom class", loc='upper right')
```

```python
plt.figure(dpi=120)
sns.boxplot(data=df_cap, x="cap-shape", y="cap-diameter", hue = "class", palette=["red", "green"])
plt.title("Mushrooms class by cap shape and cap diameter", fontweight= "bold")
plt.xlabel("Cap shape",fontweight= "bold")
plt.ylabel("Cap diameter", fontweight= "bold")
plt.xticks(rotation=90)
plt.legend(title="Mushroom class", loc='upper left')
```

#Heat map:

```python
df_color = df.groupby(["cap-color", "stem-color"])["class"].value_counts(normalize=True).unstack()
```

```python
df_color
```

```python
df_color.loc[df_color["p"].isna() & df_color["e"].isna()]
```

```python
df_color = df_color.drop("p", axis=1)
```

```python
df_color = df_color.fillna(0) #Tengo los ratios de comesitbles / venenosos + comestibles por combinacion de colores, como va de 0 a 1 lleno los que tienen Nan(es decir son todos venenosos) con 0
```

```python
df_color = df_color.unstack(0).droplevel(0, axis=1)
```

```python
df_color = df_color.fillna(-1) #Ahora lleno los Nan con -1 (un valor invalido cualquiera) para despues marcarlo en mi dataset de una forma distinta
```

```python
df_color = df_color.rename({'n': 'brown','b': 'buff','g': 'gray','r': 'green','p': 'pink','u': 'purple','e': 'red',
                 'w': 'white','y': 'yellow','l': 'blue','o': 'orange','k': 'black', "f" : "none"}, axis=0)
```

```python
df_color = df_color.rename({'n': 'brown','b': 'buff','g': 'gray','r': 'green','p': 'pink','u': 'purple','e': 'red',
                 'w': 'white','y': 'yellow','l': 'blue','o': 'orange','k': 'black'}, axis=1)
```

```python
df_color
```

```python
plt.figure(dpi=130)
cmap = sns.diverging_palette(10, 145, as_cmap = True)
cmap.set_under('#000000')
sns.heatmap(df_color, square=True, cmap=cmap,vmin=0, linewidth= .5)
plt.title("Edible/total % by cap color and stem color", fontweight= "bold")
plt.suptitle("Black means no mushroom with that combination", y=-0.07, fontsize=8)
plt.xlabel("Cap color", fontweight= "bold")
plt.ylabel("Stem color", fontweight= "bold")
```
