Veo que columnas contienen NaN y cuantos son por columna

```python
nan_columns = df.columns[df.isna().any()]
df[nan_columns].isna().sum()
```

Busco los NaN que hay en la columna `title` porque me resulto curioso que eran tan pocos

```python
df.loc[df['title'].isna()]
```

Elimino los Nan de las columnas que vi que contenian

```python
df.dropna(subset = ['title', 'parent_revision'], inplace = True)
df
```

Hago el cálculo de lo pedido

```python
df['dist_entre_ids_de_revision'] = (df.revision_id - df.parent_revision).abs()
df
```

Calculo lo pedido, veo que valor es el máximo y cual el mínimo para poder acceder al `title` del articulo, su `id`. Además se puede leer el `revision_id`	y el `parent_revision` para efectivamente chequear ese resultado. Devuelvo un df para max, y otro df para min

```python
dist_maxima = df['dist_entre_ids_de_revision'].max()
dist_maxima
```

```python
df_dist_maxima = df.loc[df['dist_entre_ids_de_revision'] == dist_maxima ]
```

```python
dist_minima = df['dist_entre_ids_de_revision'].min()
dist_minima
```

```python
df_dist_minima = df.loc[df['dist_entre_ids_de_revision'] == dist_minima ]
```
