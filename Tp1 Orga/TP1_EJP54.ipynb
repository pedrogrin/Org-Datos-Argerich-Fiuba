Veo que la columna `action` tiene un numero limitado de palabras, por lo que al pasarla a tipo category ahorro memoria

```python
df.action.unique()
```

```python
df.action = df.action.astype("category")
```

```python
df.memory_usage()
```

Me fijo cuales son los usuarios registrados y cuales no. Asi puedo descartar a los registrados y quedarme con la parte que me interesa

```python
df['registrados'] = df.contributor_ip.isna()
```

```python
df
```

```python
df = df[~df.registrados]
```

```python
df
```

Ahora con el df más limpio me fijo que no me quede nigun Nan

```python
nan_columns = df.columns[df.isna().any()]
df[nan_columns].isna().sum()
```

Ahora teniendo todos los datos que necesito ya púedo agrupar por acciones para contar la cantidad y así ver la acción más realizada por usuarios no registrados

```python
cant_action = df.groupby(['action']).agg({"item_id":["count"]}).reset_index()
cant_action.columns = cant_action.columns.droplevel(0)
cant_action
```

```python
indice_max_act = cant_action['count'].idxmax()
cant_action.loc[indice_max_act]
```
